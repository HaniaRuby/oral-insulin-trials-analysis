### Data Quality:
    - There are missing values in the ['normalized-losses', 'num-of-doors', 'horsepower', 'peak-rpm', 'bore', 'stroke' ,'price'] marked as '?'
    - The ['normalized-losses', 'horsepower', 'peak-rpm', 'price'] column is an object instead of int
    - The ['bore', 'stroke'] column is an object instead of float
    - num-of-doors, num-of-cylinders has numerical values written as string instead of int but we will leave them because they are types of cars.
    -The price column seems to have outliers

```python
#replace ? with NaN
df.replace('?', np.nan, inplace=True)
```

```python
# Check if ? are replaced
count_question_marks = df.apply(lambda x: x.eq('?').sum())
count_question_marks.sum()
```

```python
# Change numerical columns into float
df[['normalized-losses', 'horsepower', 'peak-rpm', 'price', 'bore', 'stroke']] = df[['normalized-losses', 'horsepower', 'peak-rpm', 'price', 'bore', 'stroke']].astype(np.float64)
```

```python
# Check datatypes are changed
df.dtypes
```

```python
df.describe()
```

```python
# Replace columns with missing values with the mean.
cols_missing_values = ['normalized-losses', 'horsepower', 'peak-rpm', 'bore', 'stroke' ,'price']

for col in cols_missing_values:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Replace num-of-doors with mode
most_frequent_value = df['num-of-doors'].mode()[0]

# Fill NaN values with the most frequent value
df['num-of-doors'] = df['num-of-doors'].fillna(most_frequent_value)
```

```python
# Check all nulls are filled
df.isnull().sum().sum()
```

```python
df.to_csv("cleaned_car")
```

### Visualization

```python
plt.figure(figsize=(6,4))
plt.hist(df["horsepower"], bins=3)

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("Horsepower")
```

```python
#Subplot, Markers, SuperTitle, Fontdict Example:
plt.figure(figsize=(12,4))


font1={'family': 'Tahoma', 'color':'Maroon', 'size':20}

#plot 1:
gp = df.groupby("make")['price'].mean()

plt.plot(gp.index, gp.values, marker='*')
plt.title("Average Price of Car by Make", fontdict=font1)
plt.xticks(rotation=30)
plt.show()
```
